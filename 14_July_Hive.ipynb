```python
from pyhive import hive

def create_employees_table(hive_host, hive_port, database):
    # Create a connection to Hive
    conn = hive.Connection(host=hive_host, port=hive_port, database=database)
    cursor = conn.cursor()
    
    # Execute the HiveQL query to create the table
    create_table_query = '''
    CREATE TABLE IF NOT EXISTS Employees (
        id INT,
        name STRING,
        salary FLOAT
    )
    '''
    cursor.execute(create_table_query)
    
    # Commit and close the connection
    conn.commit()
    conn.close()
    
    print("Table 'Employees' created successfully.")

if __name__ == '__main__':
    # Set the Hive host, port, and database
    hive_host = 'localhost'
    hive_port = 10000
    database = 'your_database'
    
    # Call the function to create the table
    create_employees_table(hive_host, hive_port, database)
```

```python
from pyhive import hive

def calculate_average_salary(hive_host, hive_port, database, table):
    # Create a connection to Hive
    conn = hive.Connection(host=hive_host, port=hive_port, database=database)
    cursor = conn.cursor()
    
    # Execute the HiveQL query to calculate the average salary
    query = f"SELECT AVG(salary) FROM {table}"
    cursor.execute(query)
    
    # Fetch the result
    result = cursor.fetchone()
    average_salary = result[0]
    
    # Close the connection
    conn.close()
    
    # Print the average salary
    print(f"Average salary of employees: {average_salary}")

if __name__ == '__main__':
    # Set the Hive host, port, database, and table
    hive_host = 'localhost'
    hive_port = 10000
    database = 'your_database'
    table = 'Employees'
    
    # Call the function to calculate the average salary
    calculate_average_salary(hive_host, hive_port, database, table)
```

```python
from pyhive import hive

def create_partitioned_table(hive_host, hive_port, database, table):
    # Create a connection to Hive
    conn = hive.Connection(host=hive_host, port=hive_port, database=database)
    cursor = conn.cursor()
    
    # Execute the HiveQL query to create the partitioned table
    create_table_query = f'''
    CREATE TABLE IF NOT EXISTS {table} (
        id INT,
        product STRING,
        price FLOAT
    )
    PARTITIONED BY (year INT, month INT)
    '''
    cursor.execute(create_table_query)
    
    # Close the connection
    conn.close()
    
    print(f"Partitioned table '{table}' created successfully.")

if __name__ == '__main__':
    # Set the Hive host, port, database, and table
    hive_host = 'localhost'
    hive_port = 10000
    database = 'your_database'
    table = 'Sales_Data'
    
    # Call the function to create the partitioned table
    create_partitioned_table(hive_host, hive_port, database, table)
```

```python
from pyhive import hive

def add_column_to_table(hive_host, hive_port, database, table, column_name, column_type):
    # Create a connection to Hive
    conn = hive.Connection(host=hive_host, port=hive_port, database=database)
    cursor = conn.cursor()
    
    # Execute the HiveQL query to add the new column
    alter_table_query = f"ALTER TABLE {table} ADD COLUMNS ({column_name} {column_type})"
    cursor.execute(alter_table_query)
    
    # Close the connection
    conn.close()
    
    print(f"Column '{column_name}' of type '{column_type}' added to table '{table}' successfully.")

if __name__ == '__main__':
    # Set the Hive host, port, database, table, column name, and column type
    hive_host = 'localhost'
    hive_port = 10000
    database = 'your_database'
    table = 'Employees'
    column_name = 'email'
    column_type = 'STRING'
    
    # Call the function to add the new column
    add_column_to_table(hive_host, hive_port, database, table, column_name, column_type)
```

```python
from pyhive import hive

def perform_inner_join(hive_host, hive_port, database, orders_table, customers_table, join_column):
    # Create a connection to Hive
    conn = hive.Connection(host=hive_host, port=hive_port, database=database)
    cursor = conn.cursor()
    
    # Execute the HiveQL query to perform the inner join
    join_query = f'''
    SELECT *
    FROM {orders_table} o
    INNER JOIN {customers_table} c
    ON o.{join_column} = c.{join_column}
    '''
    cursor.execute(join_query)
    
    # Fetch and print the joined records
    records = cursor.fetchall()
    for record in records:
        print(record)
    
    # Close the connection
    conn.close()

if __name__ == '__main__':
    # Set the Hive host, port, database, orders table, customers table, and join column
    hive_host = 'localhost'
    hive_port = 10000
    database = 'your_database'
    orders_table = 'Orders'
    customers_table = 'Customers'
    join_column = 'common_column'
    
    # Call the function to perform the inner join
    perform_inner_join(hive_host, hive_port, database, orders_table, customers_table, join_column)
```

```python
from pyhive import hive

def process_json_data(hive_host, hive_port, database, table):
    # Create a connection to Hive
    conn = hive.Connection(host=hive_host, port=hive_port, database=database)
    cursor = conn.cursor()
    
    # Execute the HiveQL query to process JSON data
    query = f'''
    SELECT *
    FROM {table}
    '''
    cursor.execute(query)
    
    # Fetch and print the processed JSON data
    records = cursor.fetchall()
    for record in records:
        print(record)
    
    # Close the connection
    conn.close()

if __name__ == '__main__':
    # Set the Hive host, port, database, and table
    hive_host = 'localhost'
    hive_port = 10000
    database = 'your_database'
    table = 'User_Activity_Logs'
    
    # Call the function to process JSON data
    process_json_data(hive_host, hive_port, database, table)
```
