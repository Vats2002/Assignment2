```python
def is_anagram(str1, str2):
    # Remove whitespace and convert to lowercase
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    # Check if lengths are equal
    if len(str1) != len(str2):
        return False

    # Convert strings to lists and sort them
    str1_list = sorted(list(str1))
    str2_list = sorted(list(str2))

    # Compare the sorted lists
    if str1_list == str2_list:
        return True
    else:
        return False

# Example usage
string1 = "listen"
string2 = "silent"

if is_anagram(string1, string2):
    print(f"{string1} and {string2} are anagrams.")
else:
    print(f"{string1} and {string2} are not anagrams.")
```

```python
def find_min_max(dictionary):
    if len(dictionary) == 0:
        return None, None
    
    values = list(dictionary.values())
    return min(values), max(values)

# Example usage
my_dict = {'a': 10, 'b': 5, 'c': 20}

minimum, maximum = find_min_max(my_dict)
print("Minimum value:", minimum)
print("Maximum value:", maximum)
```

```python
def calculate_average(numbers):
    if len(numbers) == 0:
        return None

    return sum(numbers) / len(numbers)

# Example usage
my_list = [10, 20, 30, 40, 50]

average = calculate_average(my_list)
print("Average:", average)
```

```python
def remove_vowels(string):
    vowels = "aeiouAEIOU"
    return ''.join(char for char in string if char not in vowels)

# Example usage
my_string = "Hello, World!"

result = remove_vowels(my_string)
print("Result:", result)
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Example usage
number = 5

result = factorial(number)
print("Factorial:", result)
```

```python
def merge_sort_merge(left, right):
    merged = []
    left_index = 0
    right_index = 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    merged.extend(left[left_index:])
    merged.extend(right[right_index:])

    return merged

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    left = merge_sort(left)
    right = merge_sort(right)

    return merge_sort_merge(left, right)

# Example usage
list1 = [5, 3, 9, 1]
list2 = [8, 2, 7, 4]

merged_list = merge_sort(list1 + list2)
print("Merged and Sorted List:", merged_list)
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Example usage
num1 = 24
num2 = 36

result = gcd(num1, num2)
print("GCD:", result)
```

```python
def find_second_largest(numbers):
    if len(numbers) < 2:
        return None

    largest = float('-inf')
    second_largest = float('-inf')

    for num in numbers:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    if second_largest == float('-inf'):
        return None

    return second_largest

# Example usage
my_list = [10, 5, 8, 20, 15]

result = find_second_largest(my_list)
if result is None:
    print("No second-largest element found.")
else:
    print("Second-largest element:", result)
```

```python
def longest_common_subsequence(str1, str2):
    m = len(str1)
    n = len(str2)

    # Create a 2D table to store the lengths of the longest common subsequences
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill the table in a bottom-up manner
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # Retrieve the longest common subsequence from the table
    lcs = ""
    i = m
    j = n
    while i > 0 and j > 0:
        if str1[i - 1] == str2[j - 1]:
            lcs = str1[i - 1] + lcs
            i -= 1
            j -= 1
        elif dp[i - 1][j] > dp[i][j - 1]:
            i -= 1
        else:
            j -= 1

    return lcs

# Example usage
string1 = "AGGTAB"
string2 = "GXTXAYB"

lcs = longest_common_subsequence(string1, string2)
print("Longest Common Subsequence:", lcs)
```
