```python
from pyhive import hive

# Specify the Hive server host and port
hive_host = '<hive_host>'
hive_port = 10000

# Establish a connection to Hive
connection = hive.Connection(host=hive_host, port=hive_port)

# Create a cursor object
cursor = connection.cursor()

# HiveQL query to create the Employees table
create_table_query = """
    CREATE TABLE Employees (
        id INT,
        name STRING,
        salary DOUBLE
    )
"""

# Execute the query to create the table
cursor.execute(create_table_query)

# Close the connection
connection.close()

print("Employees table created successfully.")
```

```python
from pyhive import hive

# Specify the Hive server host and port
hive_host = '<hive_host>'
hive_port = 10000

# Establish a connection to Hive
connection = hive.Connection(host=hive_host, port=hive_port)

# Create a cursor object
cursor = connection.cursor()

# HiveQL query to calculate the average salary
average_salary_query = """
    SELECT AVG(salary) FROM Employees
"""

# Execute the query to calculate the average salary
cursor.execute(average_salary_query)

# Fetch the average salary
result = cursor.fetchone()

# Extract the average salary from the result
average_salary = result[0]

# Print the average salary
print("Average Salary: ", average_salary)

# Close the connection
connection.close()
```

```python
from pyhive import hive

# Specify the Hive server host and port
hive_host = '<hive_host>'
hive_port = 10000

# Establish a connection to Hive
connection = hive.Connection(host=hive_host, port=hive_port)

# Create a cursor object
cursor = connection.cursor()

# HiveQL query to perform an inner join between the Orders and Customers tables
join_query = """
    SELECT *
    FROM Orders
    INNER JOIN Customers
    ON Orders.customer_id = Customers.customer_id
"""

# Execute the join query
cursor.execute(join_query)

# Fetch all the joined records
results = cursor.fetchall()

# Print the joined records
for row in results:
    print(row)

# Close the connection
connection.close()
```
